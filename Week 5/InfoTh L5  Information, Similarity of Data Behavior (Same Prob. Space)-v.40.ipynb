```python
import numpy as np
import matplotlib.pyplot as plt

Nx = 300
x = np.random.normal(loc=50, scale=10, size=Nx)

# Plot samples in sample domain
plt.figure()
plt.hist(x, bins=20)
plt.title("Sample Domain")
plt.xlabel("x")
plt.ylabel("Frequency")

# Plot samples in probability domain
counts, bins = np.histogram(x, bins=20, density=True)
plt.figure()
plt.plot(bins[1:], counts)
plt.title("Probability Domain")
plt.xlabel("x")
plt.ylabel("p(x)")

plt.show()
```

```python
import numpy as np 
import matplotlib.pyplot as plt

Ny = 500
n = 100 
p = 0.45

# Generate samples
y = np.random.binomial(n, p, size=Ny)

# Plot sample domain 
plt.figure()
plt.hist(y, bins=np.arange(n+1)-0.5)
plt.title("Sample Domain")
plt.xlabel("y")
plt.ylabel("Frequency")

# Plot probability domain
counts, bins = np.histogram(y, bins=np.arange(n+1)-0.5, density=True)
plt.figure()
plt.plot(bins[1:], counts)
plt.title("Probability Domain")
plt.xlabel("y")  
plt.ylabel("p(y)")

plt.show()
```

```python
import numpy as np
from scipy.stats import norm, binom

mu = 50 
sigma = 10
n = 100
p = 0.45

px = norm(loc=mu, scale=sigma).pdf
py = binom.pmf(np.arange(n+1), n, p)

D_KL = np.sum(px(np.arange(n+1)) * np.log(px(np.arange(n+1)) / py))

print("KL Divergence:", D_KL)
```

```python
import numpy as np
from scipy.stats import norm, binom

mu = 50
sigma = 10  
n = 100
p = 0.45

py = binom.pmf(np.arange(n+1), n, p)  
px = norm(loc=mu, scale=sigma).pdf(np.arange(n+1))

D_KL = np.sum(py * np.log(py / px))

print("KL Divergence:", D_KL)
```

```python
import numpy as np
from scipy.stats import norm, binom
import matplotlib.pyplot as plt

mu = 50
sigma = 10
n = 100
p_vals = np.arange(0.1, 1.0, 0.1)

D_KL = []
for p in p_vals:
  py = binom.pmf(np.arange(n+1), n, p)
  px = norm(loc=mu, scale=sigma).pdf(np.arange(n+1))
  D_KL.append(np.sum(py * np.log(py / px)))

plt.plot(p_vals, D_KL)
plt.xlabel('p')
plt.ylabel('D_KL(X || Y)')
plt.title('KL Divergence vs Binomial p')
plt.show()
```

```python
import numpy as np
from scipy.stats import norm, binom 
import matplotlib.pyplot as plt

mu = 50
sigma = 10
n = 100
p_vals = np.arange(0.1, 1.0, 0.1)

dist = []
for p in p_vals:
  px = norm(loc=mu, scale=sigma).pdf(np.arange(n+1))
  py = binom.pmf(np.arange(n+1), n, p)
  
  # Calculate Euclidean distance   
  dist.append(np.linalg.norm(px - py))

plt.plot(p_vals, dist)
plt.xlabel('p')  
plt.ylabel('Euclidean Distance')
plt.title('Distance between PDFs')
plt.show()
```

```python
import numpy as np
from scipy.stats import norm, binom
import matplotlib.pyplot as plt

num_trials = 10
mu = 50
sigma = 10
n = 100
p_vals = np.arange(0.1, 1.0, 0.1)

fig, axs = plt.subplots(2, num_trials, figsize=(15,5))

for t in range(num_trials):
  
  D_KL = []
  dist = []
  
  for p in p_vals:
    x = np.random.normal(loc=mu, scale=sigma, size=n) 
    y = np.random.binomial(n, p, size=n)
    
    px = norm(loc=mu, scale=sigma).pdf(x)
    py = binom.pmf(y, n, p)
    
    D_KL.append(np.sum(px * np.log(px / py)))
    dist.append(np.linalg.norm(px - py))

  axs[0, t].plot(p_vals, D_KL)
  axs[0, t].set_title('Trial %d KL Divergence' % (t+1))
  
  axs[1, t].plot(p_vals, dist)
  axs[1, t].set_title('Trial %d Euclidean Distance' % (t+1))
  
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Generate samples
Nx = 1000
X = np.random.normal(size=(Nx,2)) 

Ny = 2000
Y = np.random.uniform(low=0, high=100, size=(Ny,2))

# Plot X samples
plt.figure()
plt.hist(X[:,0], bins=20)
plt.title("X1 Sample Domain")

plt.figure() 
plt.hist(X[:,1], bins=20)
plt.title("X2 Sample Domain")

# Plot Y samples
plt.figure()
plt.hist(Y[:,0], bins=20)
plt.title("Y1 Sample Domain")

plt.figure()
plt.hist(Y[:,1], bins=20) 
plt.title("Y2 Sample Domain")

# Plot X points
plt.figure()
plt.plot(X[:,0], X[:,1], '.')
plt.title("X1 vs X2")

# Plot Y points
plt.figure() 
plt.plot(Y[:,0], Y[:,1], '.')
plt.title("Y1 vs Y2")


# Plot X probability
nbins = 20
H, xedges, yedges = np.histogram2d(X[:,0], X[:,1], bins=nbins) 
xidx = np.argsort(xedges)
yidx = np.argsort(yedges)
H = H / Nx
plt.figure()  
plt.pcolormesh(xedges[xidx], yedges[yidx], H, cmap='Blues')
plt.title("X Probability")

# Plot Y probability
H, xedges, yedges = np.histogram2d(Y[:,0], Y[:,1], bins=nbins)
H = H / Ny 
plt.figure()
plt.pcolormesh(xedges[xidx], yedges[yidx], H, cmap='Reds')
plt.title("Y Probability")

# Take equal samples from X and Y
X_subset = np.array(X[np.random.choice(Nx, size=1000)]).ravel() 
Y_subset = np.array(Y[np.random.choice(Ny, size=1000)]).ravel()

# Plot XY probability
H, xedges, yedges = np.histogram2d(X_subset, Y_subset, bins=nbins)
H = H / 1000  
plt.figure()
plt.pcolormesh(xedges[xidx], yedges[yidx], H, cmap='gray')
plt.title("XY Probability")

# Calculate KL divergence
px = norm(loc=0, scale=1).pdf(X).sum(axis=1)  
py = np.full(Nx, 1/100) 

D_KL_XY = np.sum(px * np.log(px / py))
D_KL_YX = np.sum(py * np.log(py / px))

print("D_KL(X || Y):", D_KL_XY) 
print("D_KL(Y || X):", D_KL_YX)

plt.show()
```
